```python
import numpy as np
```

```python
np.random.seed(1)
```

```python
data = np.random.randint(1,6,size=(100,4))
```

```python
data
```

```python
min = data.min()
```

```python
max = data.max()
```

```python
print(min)
```

```python
print(max)
```

```python
normalize = (data-min)/(max-min)
```

```python
employee_avg = normalize.mean(axis=1)
```

```python
employee_avg
```

```python
company_avg = normalize.mean()
```

```python
company_avg
```

```python
highest_employee = np.where(employee_avg>company_avg)[0]
```

```python
print(highest_employee)
```

```python
print(company_avg)
```
