```python
import numpy as np

np.random.seed(42)

heart_rate = np.random.randint(55, 110, size=(60, 24)).astype(float)

# introduce missing values
heart_rate[2, 5] = np.nan
heart_rate[18, 12] = np.nan
heart_rate[40, 20] = np.nan
```

```python
heart_rate
```

```python
hourly_avg_heart_rate = np.nanmean(heart_rate, axis=0)
hourly_avg_heart_rate
```

```python
abnormal_mask = (heart_rate < 60) | (heart_rate > 100)
abnormal_mask
```

```python
filled_heart_rate = heart_rate.copy()

for patient in range(filled_heart_rate.shape[0]):
    for hour in range(1, filled_heart_rate.shape[1]):
        if np.isnan(filled_heart_rate[patient, hour]):
            filled_heart_rate[patient, hour] = filled_heart_rate[patient, hour - 1]
```
