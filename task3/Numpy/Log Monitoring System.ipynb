```python
import numpy as np
```

```python
np.random.seed(42)
requests = np.random.randint(50, 500, size=1440)
```

```python
hourly_traffic = requests.reshape(24, 60).sum(axis=1)
```

```python
peak_hour = np.argmax(hourly_traffic)
```

```python
print("Hourly traffic:", hourly_traffic)
print("Peak traffic hour:", peak_hour)
```

```python
window_size = 10
moving_avg = np.convolve(requests, np.ones(window_size)/window_size, mode='valid')

print("Smoothed data (first 10 values):", moving_avg[:10])
```

```python
threshold = requests.mean() + 2 * requests.std()
spikes = np.where(requests > threshold)[0]

print("Spike indices:", spikes[:10])
```
