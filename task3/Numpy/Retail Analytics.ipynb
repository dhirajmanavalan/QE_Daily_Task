```python
import numpy as np
```

```python
np.random.seed(1)
```

```python
data = np.random.randint(-25,100,size=(30,5))
```

```python
data
```

```python
# Im replacing here negative values
data[data<0] = 0
```

```python
data
```

```python
# here im calculating for only 28 days
weeks = 4
days = 7
sales = data[:weeks*days]
```

```python
sales
```

```python
# here im reshaping based on week , days, products
products = 5
weekly_sales = sales.reshape(weeks,days,products)
```

```python
weekly_sales
```

```python
week_total = weekly_sales.sum(axis=1)
```

```python
week_total
```

```python
avg = week_total.mean(axis=0)
```

```python
# average weekly sales
avg
```

```python
# highest average weekly sales index
highestavg = np.argmax(avg)
```

```python
print(highestavg + 1)
```
