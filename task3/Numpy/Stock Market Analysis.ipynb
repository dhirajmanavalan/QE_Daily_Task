```python
import numpy as np
```

```python
np.random.seed(42)
prices = np.random.randint(100, 500, size=(90, 10))
```

```python
daily_returns = (prices[1:] - prices[:-1]) / prices[:-1]
```

```python
volatility = np.std(daily_returns, axis=0)
```

```python
highest_risk_stock = np.argmax(volatility)
```

```python
print("Volatility per stock:", volatility)
print("Highest risk stock index:", highest_risk_stock)
print("Highest volatility value:", volatility[highest_risk_stock])
```
