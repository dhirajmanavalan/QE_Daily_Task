```python
mean_temp = np.mean(s)
std_temp = np.std(s)
```

```python
lower_bound = mean_temp - 2 * std_temp
upper_bound = mean_temp + 2 * std_temp
```

```python
temps_cleaned = s.copy()
temps_cleaned[(s < lower_bound) | (s > upper_bound)] = mean_temp
```
