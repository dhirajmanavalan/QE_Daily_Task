```python
import pandas as pd
```

```python
data = {
    "account_id": [1001, 1001, 1002, 1003, 1002, 1001, 1003, 1004, 1004],
    "transaction_date": [
        "2026-01-10", "2026-01-10", "2026-01-10",
        "2026-01-11", "2026-01-11", "2026-01-11",
        "2026-01-12", "2026-01-12", "2026-01-12"
    ],
    "amount": [5000, -2000, 15000, 3000, 22000, -1000, 40000, 800, 60000],
    "transaction_type": [
        "credit", "debit", "credit",
        "credit", "credit", "debit",
        "credit", "debit", "credit"
    ]
}
```

```python
df = pd.DataFrame(data)
df["transaction_date"] = pd.to_datetime(df["transaction_date"])
```

```python
daily_totals = (
    df.groupby(["account_id", "transaction_date"])["amount"]
      .sum()
      .reset_index()
)

daily_totals
```

```python
account_volume = (
    df.groupby("account_id")["amount"]
      .sum()
      .sort_values(ascending=False)
)

account_volume
```

```python
threshold = account_volume.quantile(0.90)
high_volume_accounts = account_volume[account_volume > threshold]

high_volume_accounts
```

```python
mean_amt = df["amount"].mean()
std_amt = df["amount"].std()

fraud_threshold = mean_amt + 3 * std_amt

potential_fraud = df[df["amount"] > fraud_threshold]
potential_fraud
```
