```python
import pandas as pd
```

```python
data = {
    "patient_id": [201, 202, 203, 201, 202, 203, 204, 204],
    "timestamp": [
        "2026-01-15 08:00", "2026-01-15 08:00", "2026-01-15 08:00",
        "2026-01-15 12:00", "2026-01-15 12:00", "2026-01-15 12:00",
        "2026-01-15 08:00", "2026-01-15 12:00"
    ],
    "heart_rate": [78, 82, 90, 110, 85, 140, 75, 130],
    "ward": ["General", "General", "ICU", "General", "General", "ICU", "ICU", "ICU"]
}
```

```python
df = pd.DataFrame(data)
df["timestamp"] = pd.to_datetime(df["timestamp"])
```

```python
avg_hr_per_ward = (
    df.groupby("ward")["heart_rate"]
      .mean()
)

avg_hr_per_ward
```

```python
df = df.sort_values(["patient_id", "timestamp"])

df["hr_change"] = (
    df.groupby("patient_id")["heart_rate"]
      .diff()
)

sudden_spikes = df[df["hr_change"] > 30]
sudden_spikes
```

```python
critical_cases = df[df["heart_rate"] > 120]
critical_cases
```

```python
icu_critical = df[
    (df["ward"] == "ICU") & (df["heart_rate"] > 120)
]
icu_critical
```
