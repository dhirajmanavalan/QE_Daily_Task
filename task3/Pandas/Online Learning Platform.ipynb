```python
import pandas as pd
```

```python
data = {
    "student_id": [101, 102, 103, 101, 104, 102, 105],
    "course_id": ["C1", "C1", "C2", "C2", "C3", "C3", "C1"],
    "login_date": [
        "2026-01-10", "2026-01-12", "2026-01-05",
        "2026-01-14", "2026-01-03", "2026-01-15",
        "2026-01-01"],
    "minutes_spent": [30, 45, 60, 40, 20, 50, 25]
}
```

```python
s = pd.DataFrame(data)
s['login_date'] = pd.to_datetime(s['login_date'])
s
```

```python
latest_date = s["login_date"].max()
cutoff_date = latest_date - pd.Timedelta(days=7)

inactive_students = (
    s.groupby("student_id")["login_date"]
      .max()
      .loc[lambda x: x < cutoff_date]
      .index
)

print(inactive_students)
```

```python
avg = (s.groupby('course_id')['minutes_spent'].mean())
avg
```

```python
top_course = (s.groupby('course_id')['minutes_spent'].sum().sort_values(ascending=False).head(3))
top_course
```
