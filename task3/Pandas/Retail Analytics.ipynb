```python
import pandas as pd
```

```python
values = {
    "order_id": [1, 2, 3, 4, 5, 6],
    "customer_id": [101, 102, 101, 103, 104, 102],
    "order_date": ["2024-01-10", "2024-01-15", "2024-02-05","2024-02-20", "2024-03-10", "2024-03-15"],
    "order_amount": [250, 400, 300, 500, 450, 350]
}
```

```python
data = pd.DataFrame(values)
```

```python
data
```

```python
data['order_date'] = pd.to_datetime(data["order_date"])
```

```python
data.info()
```

```python
month_revenue = (data.groupby(data["order_date"].dt.to_period("M"))["order_amount"].sum())
```

```python
month_revenue
```

```python
month_Customers = (data.groupby(data["order_date"].dt.to_period("M"))["customer_id"].nunique())
```

```python
month_Customers
```

```python
top_month = month_Customers.idxmax()
```

```python
top_month
```
