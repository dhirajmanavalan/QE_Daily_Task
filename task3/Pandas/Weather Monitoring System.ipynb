```python
import pandas as pd
```

```python
data = {
    "date": [
        "2026-01-01", "2026-01-02", "2026-01-03",
        "2026-01-04", "2026-01-05", "2026-01-06"
    ],
    "city": [
        "Chennai", "Chennai", "Chennai",
        "Bangalore", "Bangalore", "Bangalore"
    ],
    "temperature": [30, 32, 31, 25, 26, 27],
    "humidity": [70, 72, 71, 65, 67, 70]
}
```

```python
s=pd.DataFrame(data)
```

```python
s
```

```python
s["date"] = pd.to_datetime(s["date"])
```

```python
chennai = s[s['city']=='Chennai']
```

```python
chennai
```

```python
chennai['rolling'] = (chennai['temperature'].rolling(window=3).mean())
```

```python
chennai
```

```python
chennai['humidity_increased'] = chennai['humidity']>chennai['humidity'].shift(1)
```
